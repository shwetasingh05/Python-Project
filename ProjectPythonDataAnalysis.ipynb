```python
#OBJECTIVE 1: HEALTH BEHAVIOR TRENDS OVER TIME 
import seaborn as sns
import matplotlib.pyplot as plt
# Filter relevant health classes
trend_classes = ['Obesity / Weight Status', 'Physical Activity', 'Fruits and Vegetables']
trend_df = nutrition[nutrition['Class'].isin(trend_classes)]
# Group and average
trend_summary = trend_df.groupby(['YearStart', 'Class'])['Data_Value'].mean().reset_index()
plt.figure(figsize=(10, 5))
sns.lineplot(data=trend_summary, x='YearStart', y='Data_Value', hue='Class', marker='o')
plt.title("Health Behavior Trends Over Time")
plt.xlabel("Year")
plt.ylabel("Average Data Value (%)")
plt.legend(title="Health Class")
plt.tight_layout()
plt.show()
```

```python
# OBJECTIVE 2: STATE-LEVEL COMPARISON
heat_df = nutrition[(nutrition['Class'] == 'Obesity / Weight Status') & (nutrition['Income'].notna())]
heatmap_data = heat_df.pivot_table(index='LocationDesc', columns='Income', values='Data_Value', aggfunc='mean')
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt=".1f", cmap='Purples')
plt.title("Obesity Rate Heatmap by State and Income")
plt.xlabel("Income Level")
plt.ylabel("State")
plt.tight_layout()
plt.show()
```

```python
# OBJECTIVE 3: CORRELATION BEWTEEN HEALTH BEHAVIORS 
class_counts = nutrition[nutrition['Class'].isin(['Obesity / Weight Status', 'Physical Activity', 'Fruits and Vegetables'])]['Class'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Health Behavior Distribution in Dataset")
plt.tight_layout()
plt.show()
```

```python
# OBJECTIVE 4: TOP 10 STATES BY AVERAGE OBESITY RATE
latest_year = nutrition['YearStart'].max()
obesity_df = nutrition[(nutrition['Class'] == 'Obesity / Weight Status') &(nutrition['YearStart'] == latest_year)]
state_avg = obesity_df.groupby('LocationDesc')['Data_Value'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=state_avg.values, y=state_avg.index, color='pink')
plt.title(f"Top 10 States by Obesity Rate ({latest_year})")
plt.xlabel("Average Obesity Rate (%)")
plt.ylabel("State")
plt.tight_layout()
plt.show()
```

```python
#OBJECTIVE 5: OBESITY vs PHYSICAL ACTIVITY(BY STATE)
scatter_df = nutrition[(nutrition['Class'].isin(['Obesity / Weight Status', 'Physical Activity'])) &(nutrition['YearStart'] == nutrition['YearStart'].max())]
scatter_pivot = scatter_df.pivot_table(index='LocationDesc', columns='Class', values='Data_Value', aggfunc='mean').dropna()
scatter_pivot = scatter_pivot.rename(columns={'Obesity / Weight Status': 'Obesity','Physical Activity': 'PhysicalActivity'})
plt.figure(figsize=(7, 5))
sns.scatterplot(data=scatter_pivot, x='PhysicalActivity', y='Obesity', s=80, color='teal')
plt.title('Obesity vs Physical Activity by State')
plt.xlabel('Physical Activity (%)')
plt.ylabel('Obesity Rate (%)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
